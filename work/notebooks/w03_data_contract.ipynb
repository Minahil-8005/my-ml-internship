# ML-04 — Search Intelligence Data Contract

This notebook establishes the formal data contract for the Page Refresh / Re-scoring lane. It verifies data grain, specifies features vs. targets, checks null availability constraints, and demonstrates data leakage prevention.

## 1. Unit of analysis + time window

*One row = one what, over which dates? State it, then verify it below.*

- **Unit of Analysis (Grain):** One row represents **one URL / Page (`page_url`)** aggregated over the observation period.
- **Time Window:** Mid-panel observation month: **`2026-03-01` to `2026-03-31`** (Treating the final month June 2026 as a sealed test window).
- **Target Decision:** Predicting page traffic decay to select the top 50 URLs for monthly content refreshes.

```python
# Setup Hugging Face Token & Load Mid-panel Dataset (March 2026)
import os
import pandas as pd
import numpy as np
from datasets import load_dataset

# Authenticate via HF_TOKEN (stored in Colab Secrets)
hf_token = os.environ.get('HF_TOKEN')

# Load warehouse mid-panel sample
dataset = load_dataset('FlyRank/internship-warehouse', split='train', token=hf_token)
df = pd.DataFrame(dataset)

# Filter for mid-panel month: 2026-03
df['date'] = pd.to_datetime(df['date'])
march_df = df[(df['date'] >= '2026-03-01') & (df['date'] <= '2026-03-31')].copy()
print(f"Loaded March 2026 mid-panel data: {march_df.shape[0]} rows")
```

## 2. Fields: feature / label / context / excluded

*Sort every field you plan to touch into these four buckets. Excluded needs a why.*

- **Feature Bucket (Knowable at Decision Moment):**
  - `clicks_past_30d`: Organic clicks accumulated over the preceding 30-day window.
  - `impressions_past_30d`: Total search visibility impressions in the preceding 30 days.
  - `ctr_past_30d`: Historical click-through rate calculated prior to decision time.
  - `position_mean_30d`: Mean SERP ranking position over the previous 30 days.
  - `days_since_last_update`: Total days elapsed since the page content was last modified.

- **Label / Target Bucket:**
  - `is_declining` (Boolean): True if organic clicks dropped by >20% relative to the baseline period.

- **Context Bucket:**
  - `page_url`: Unique page resource identifier.
  - `category`: Structural content taxonomy classification.

- **Excluded Bucket & Reason:**
  - `future_clicks_next_30d` & `next_month_ctr`: **Excluded due to Data Leakage.** These metrics occur in the future relative to the decision timestamp and cannot be known when ranking pages for a refresh slot.

```python
# Verify field bucketing and missing value counts
feature_cols = ['clicks_past_30d', 'impressions_past_30d', 'ctr_past_30d', 'position_mean_30d', 'days_since_last_update']
label_col = 'is_declining'

print("Feature Availability Summary:")
print(march_df[feature_cols].isnull().sum())
print(f"\nLabel Distribution ('{label_col}'):")
print(march_df[label_col].value_counts(dropna=False))
```

## 3. Verify it with queries (grain, counts, missing values, windows)

*Every claim above gets a query cell here. A contract claim without a query next to it is a guess.*

```python
# Query 1: Verify Grain (1 Row = 1 Page URL in aggregated window)
page_counts = march_df.groupby('page_url').size()
duplicate_pages = (page_counts > 1).sum()
print(f"Query 1 [Grain Check]: Duplicate page_urls found = {duplicate_pages} (Expected 0 for 1 row per URL)")

# Query 2: Row Count & Date Span Verification
min_date = march_df['date'].min().strftime('%Y-%m-%d')
max_date = march_df['date'].max().strftime('%Y-%m-%d')
total_rows = len(march_df)
print(f"Query 2 [Counts & Span]: Total Rows = {total_rows}, Date Range = {min_date} to {max_date}")

# Query 3: Availability Check using IS TRUE (Handling Nil / Null Flags)
# Filtering boolean flags with IS TRUE explicitly to avoid dropping nil rows as false
if 'has_gsc_data' in march_df.columns:
    surviving_rows = (march_df['has_gsc_data'] == True).sum()
else:
    surviving_rows = march_df['clicks_past_30d'].notnull().sum()
print(f"Query 3 [Availability]: Surviving rows using IS TRUE filter = {surviving_rows} / {total_rows}")
```

```python
# Step 4: The Leakage Trap Experiment (Deliberate Leak & Recovery)
from sklearn.metrics import precision_score

# 1. Base Feature Prediction (Honest Feature Frame)
march_df['honest_score'] = march_df['clicks_past_30d'] * 0.4 + march_df['ctr_past_30d'] * 0.6
top_50_honest = march_df.sort_values(by='honest_score', ascending=False).head(50)
honest_precision = precision_score(top_50_honest[label_col], [True]*50, zero_division=0)
print(f"Honest Model Precision@50: {honest_precision:.2f}")

# 2. Injecting Leaked Feature (Future Metric Derived Column)
march_df['leaked_future_clicks'] = march_df[label_col].map({True: 0, False: 1000})
top_50_leaked = march_df.sort_values(by='leaked_future_clicks', ascending=True).head(50)
leaked_precision = precision_score(top_50_leaked[label_col], [True]*50, zero_division=0)
print(f"[TRAP] Leaked Model Precision@50: {leaked_precision:.2f} (Artificially perfect!)")

# 3. Remove Leaked Column & Restore Honest State
march_df.drop(columns=['leaked_future_clicks'], inplace=True)
print("Successfully removed leaked feature column. Honest contract preserved.")
```

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

1. **GSC-Only Early Rows:** Historical rows prior to full GA4 sync lack revenue and conversion telemetry, meaning predictions reflect traffic decay rather than direct business ROI.
2. **Unbalanced URL History:** Newly published URLs possess less than 30 days of historical baseline data, resulting in noisier trend signals compared to established pages.
3. **External SERP Shifts:** Organic traffic drops caused by Google Core Algorithm updates or competitor activity are unobserved in internal warehouse logs, providing directional decision support rather than causal certainty.
