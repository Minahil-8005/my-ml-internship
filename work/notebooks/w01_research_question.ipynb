## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Total number of pages and overall declining rate
total_pages = df.shape[0]
declining_rate = df["trend_direction"].str.lower().eq("down").mean() * 100

# 2. Number of pages that are highly visible but haven't been updated in 6 months (180 days)
stale_and_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)].shape[0]

print(f"Total Pages in Dataset: {total_pages}")
print(f"Overall Content Declining Rate: {declining_rate:.1f}%")
print(f"High-risk Stale Pages (>= 180 days old & >= 500 impressions): {stale_and_visible}")
```
