## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
Task Type: Binary Classification / Scoring

Explanation: I am framing this as a binary classification task. The system will assign a score/probability to each page indicating whether it is at high risk of a traffic decline (1 for declining, 0 for stable/growing).
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
Target Column: is_declining_label

Definition: A binary proxy indicator derived from trend_direction. If the directional trend of the search data is "down", the target label is 1 (declining), otherwise 0.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
Success Metric: Precision@K (specifically Precision@50)

Why: Since human resources are limited, the content team can only review a specific number of pages (e.g., top 50 high-risk pages) per week. We need to maximize the percentage of genuinely declining pages within that top batch to avoid wasting their time.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
import numpy as np

# Load starter data
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create the target column
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Select columns that define our unit of analysis and target snapshot
sample_view = df[["page_id", "days_since_last_update", "impressions_90d", "avg_position", "ctr", "is_declining_label"]].head(5)

print("--- Unit of Analysis: 1 Row = 1 Unique Page ---")
print(f"Dataset contains {df.shape[0]} rows (pages) and {df.shape[1]} columns.\n")
print(sample_view)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
Fixed rules (like checking if a page hasn't been updated in 180 days) are too rigid. They treat a page with a slight drop in impressions the same as a page with a catastrophic crash. Machine Learning beats a fixed rule because it learns complex, non-linear interactions between multiple features—such as combining a high position with a dropping CTR and a sudden drop in impression volume—allowing for dynamic and highly accurate risk prioritizing.
```
