# Week 2: Problem Framing Assignment

## Section 1: Lane & ML Type Choice
- **Chosen Lane:** Page Refresh / Re-scoring Lane
- **ML Type:** Supervised Learning
- **Reason:** My lane is supervised because we already have historical performance data with target labels (e.g., clicks declining or rising) to train the model.

## Section 2: Unit of Analysis
- **1 Row Definition:** 1 row = 1 URL / 1 Page from the Google Search Console dataset.

## Section 3: Core Problem Framing
- **Business Goal / Decision:** Deciding which top 50 pages get our limited content update/refresh slots this month.
- **ML Task Type:** Scoring / Ranking
- **Target Variable:** `is_declining` (Boolean flag indicating if organic traffic is dropping).
- **Success Metric:** `Precision@50` (Out of top 50 predicted pages, how many actually needed a refresh).
- **Action:** Send the top 50 ranked pages to the content team for rewriting.

## Section 4: Data Quality & Null Checks
- **Data Quality:** Verified missing values in the dataset.
- **Null Handling:** Boolean flags contain `nil` values, so filters should explicitly use `is True` instead of `= False`.
